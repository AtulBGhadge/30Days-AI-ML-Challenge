```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('../data/Iris.csv')  

print(data.head())
```

```python
data.rename(columns={
    'SepalLengthCm': 'sepal_length',
    'SepalWidthCm': 'sepal_width',
    'PetalLengthCm': 'petal_length',
    'PetalWidthCm': 'petal_width',
    'Species': 'species'
}, inplace=True)

print(data.head())
```

```python
sns.set(style="whitegrid", palette="muted")
```

```python
species_counts = data['species'].value_counts()
plt.figure(figsize=(6,6))
plt.pie(species_counts, labels=species_counts.index, autopct='%1.1f%%',
        startangle=90, shadow=True, colors=sns.color_palette('pastel'))
plt.title('Iris Species Distribution')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(data=data, x='sepal_length', y='sepal_width',
                hue='species', style='species', s=100)
plt.title('Sepal Length vs Sepal Width by Species')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.legend(title='Species')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,6))
for column in ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']:
    plt.plot(data[column], label=column, marker='o')
plt.title('Feature Trends Across Samples')
plt.xlabel('Sample Index')
plt.ylabel('Feature Value (cm)')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
avg_features = data.groupby('species').mean()
avg_features.plot(kind='bar', figsize=(8,6), color=sns.color_palette('bright'))
plt.title('Average Feature Values per Species')
plt.ylabel('Mean Value (cm)')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.histplot(data['sepal_length'], color='skyblue', kde=True, label='Sepal Length')
sns.histplot(data['petal_length'], color='orange', kde=True, label='Petal Length')
plt.title('Histogram of Sepal and Petal Length')
plt.xlabel('Length (cm)')
plt.ylabel('Count')
plt.legend()
plt.tight_layout()
plt.show()
```
